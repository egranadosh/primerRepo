```python
# Sección Listas y Duplas

"""
    SECCION LISTAS Y TUPLAS
"""

frutas = ['manzana','pera','naranjas']

print(f'Lista original: {frutas}')

frutas.extend(['fresa','mango'])
print(f'Lista 2: {frutas}')

frutas.insert(1,'uvas')
print(f'Lista 3: {frutas}')
```
